```python
def im_age(col):
    age=col[0]
    pclass=col[1]

    if pd.isnull(age):

        if pclass==1:
            return 37

        elif pclass==2:
            return 27

        else: 
            return 23

    else:
        return age
```

```python
df["Age"] = df[["Age","Pclass"]].apply(im_age , axis=1)
```

```python
sns.heatmap(df.isnull(),cbar=True,cmap="viridis")
```

```python
df.drop("Cabin",axis=1,inplace=True)
```

```python
X = df[['PassengerId','Pclass', 'Age', 'SibSp', 'Parch', 'Fare',
       'male', 'Q', 'S']]
```

```python
X
```

```python
y = df[["Survived"]]
```

```python
y
```

### Creating Decission Tree Model 

```python
from sklearn.tree import DecisionTreeRegressor

modelDTR = DecisionTreeRegressor(random_state=0,max_depth=10)

modelDTR.fit(X,y)
```

```python
from sklearn.tree import export_text

Text_exported = export_text(modelDTR)

print(Text_exported)
```

```python
from sklearn.tree import plot_tree

fig = plt.figure(figsize=(40,20) , dpi= 100)

visualization = plot_tree(modelDTR,feature_names=df.columns)

print(visualization)
```

```python
modelDTR.predict([[5,	3	,35.0,	0	,0,	8.0500,	1,	0,	1]])
```

### Calculating the Accuracy and Error  Of the Dicission Tree Model

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from math import sqrt

r2_score = r2_score(y,modelDTR.predict(X))
MAE = mean_absolute_error(y,modelDTR.predict(X))
MSE = mean_squared_error(y,modelDTR.predict(X))
RMSE = sqrt(MSE)

print("r2_score : ",r2_score)
print("MAE      : ",MAE)
print("MSE      : ",MSE)
print("RMSE     : ",RMSE)
```

### Creating Decission Linear Regression Model 

```python
from sklearn.model_selection import train_test_split

Xtrain , Xtest, ytrain , ytest = train_test_split(X,y,random_state=0)

```

```python
from sklearn.linear_model import LinearRegression

modelLR = LinearRegression()

modelLR.fit(Xtrain,ytrain)
```

```python
modelLR.predict([[5,	3	,35.0,	0	,0,	8.0500,	1,	0,	1]])
```

### Calculating the Accuracy and Error  Of the Linear Regression Model

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from math import sqrt

r2_score = r2_score(y,modelLR.predict(X))
MAE = mean_absolute_error(y,modelLR.predict(X))
MSE = mean_squared_error(y,modelLR.predict(X))
RMSE = sqrt(MSE)

print("r2_score : ",r2_score)
print("MAE      : ",MAE)
print("MSE      : ",MSE)
print("RMSE     : ",RMSE)
```

### Creating Decission Polynomial Regression Model 

```python
from sklearn.preprocessing import PolynomialFeatures

modelPF = PolynomialFeatures(degree = 9)

model_Poly = modelPF.fit_transform(X,y)

model_Lin_Poly = modelLR.fit(model_Poly,y)


```

### Calculating the Accuracy and Error  Of the Polynomial Regression  Model

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from math import sqrt

r2_score = r2_score(y,model_Lin_Poly.predict(model_Poly))
MAE = mean_absolute_error(y,model_Lin_Poly.predict(model_Poly))
MSE = mean_squared_error(y,model_Lin_Poly.predict(model_Poly))
RMSE = sqrt(MSE)

print("r2_score : ",r2_score)
print("MAE      : ",MAE)
print("MSE      : ",MSE)
print("RMSE     : ",RMSE)
```
